* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ – 3.

```python
import matplotlib.pyplot as plt
import numpy as np
import random
from time import time

import functions as func1
import functions_vectorized as func2

%matplotlib inline

time_non_vectorized = []
time_vectorized = []

data_size = np.array([0] * 30)
data_size[0] = 4500
for i in range(1, 30):
    data_size[i] = data_size[0] + 500 * i

for i in data_size:
    matr1 = [0.0] * i
    matr1 = [matr1] * i
    matr2 = np.array([[0.0] * i] * i)
    for j in range(i):
        matr1[j][j] = random.uniform(0, 1)
        matr2[j][j] = matr1[j][j]
    time_now = time()
    func2.prod_non_zero_diag(matr2)
    time_vectorized.append(time() - time_now)
    time_now = time()
    func1.prod_non_zero_diag(matr1)
    time_non_vectorized.append(time() - time_now)


f, ax = plt.subplots()

ax.plot(data_size, time_non_vectorized, label=u"Невекторизованная")
ax.plot(data_size, time_vectorized, label=u"Векторизованная")
ax.set_title(u"Хороший график")
ax.set_xlabel(u"Длина массива")
ax.set_ylabel(u"Время (с)")
ax.grid()
ax.legend()

f.show()         
```

* __Задача 5__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ – (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
import matplotlib.pyplot as plt
import numpy as np
import random
from time import time

import functions as func1
import functions_vectorized as func2

%matplotlib inline

time_non_vectorized = []
time_vectorized = []

data_size = np.array([0] * 30)
data_size[0] = 100000
for i in range(1, 30):
    data_size[i] = data_size[0] + 100000 * i

    
l1 = [0]
ar1 = np.array([0])
for i in data_size:
    for j in range(len(l1), i):
        l1.append(random.randint(0, 100000))
    ar1 = np.array(l1);
    time_now = time()
    func2.run_length_encoding(ar1)
    time_vectorized.append(time() - time_now)
    time_now = time()
    func1.run_length_encoding(l1)
    time_non_vectorized.append(time() - time_now)


f, ax = plt.subplots()

ax.plot(data_size, time_non_vectorized, label=u"Невекторизованная")
ax.plot(data_size, time_vectorized, label=u"Векторизованная")
ax.set_title(u"Хороший график")
ax.set_xlabel(u"Длина массива")
ax.set_ylabel(u"Время (с)")
ax.grid()
ax.legend()

f.show()
```
